### Выведем график распределения количества рейсов по моделям самолётов

```python
sns.set_style("darkgrid")
sns.set_context(font_scale=1.15)

data_1_sort = data_1.sort_values(by='flights_amount', ascending=False)

data_1_sort.plot(x='model', kind='bar', grid=True, figsize=(14,6))
plt.title('График распределения количества рейсов по моделям самолётов', y=1.01)
plt.xlabel('')
plt.ylabel('Количество рейсов')
plt.legend(['Количество рейсов'],loc='upper right')

vals = data_1_sort.flights_amount
labels = data_1_sort.model

fig, ax = plt.subplots(figsize=(6.5,6.5))
ax.pie(vals, labels=labels, wedgeprops=dict(width=0.6), autopct='%1.1f%%')
ax.axis("equal")
plt.title('График распределения рейсов в процентном соотношении')

plt.show()
```

### Построим график городов и количества рейсов

```python
data_2.sort_values(by='average_flights',ascending=True).plot(x='city', kind='barh', grid=True, figsize=(14,30),legend=False, align='center')
plt.title('График городов и среднего количества рейсов в день')
plt.ylabel('')
plt.xlabel('Количество рейсов')
plt.show()
```

### Вывод

### Выберем топ-10 городов по количеству рейсов и нанесём их на график для наглядности

```python
data_2_top10=data_2.sort_values(by='average_flights',ascending=False)[:10].copy()
data_2_top10
```

#### Построим график для наглядности

```python
data_2_top10.plot(x='city', kind='bar', grid=True, figsize=(14,6))
plt.title('ТОП-10 городов по количеству рейсов', y=1.01)
plt.xlabel('')
plt.ylabel('Количество рейсов')
plt.legend(['Количество рейсов'],loc='upper right')
plt.show()
```

### Вывод

   Здесь мы отчётливо видим, что Москва безусловный лидер по количеству рейсов. А так же это говорит и о том, что именно в Москве чаще всего проводятся крупнейшие культурные фестивали, не смотря на то,что Санкт-Петербург считается культруной столицей!

### Посчитем процентное распределение рейсов ТОП-10 и другие города

#### Составим таблицу и определим в одну группы остальные города

```python
data_2pie = data_2.sort_values(by='average_flights',ascending=False).copy()
data_2pie
```

```python
k=[]
for i in list(data_2pie.city):
    if i in list(data_2_top10.city):
        y = i
        k.append(y)
    else:
        y = 'Другие'
        k.append(y)
data_2pie['grup']=k
```

```python
data_2pie
```

```python
data_2pie_t=data_2pie.groupby('grup').sum().sort_values(by='average_flights',ascending=False).astype('int')
data_2pie_t=data_2pie_t.reset_index()
data_2pie_t
```

```python
vals = data_2pie_t.average_flights
labels = data_2pie_t.grup

fig, ax = plt.subplots(figsize=(6.5,6.5))
ax.pie(vals, labels=labels, wedgeprops=dict(width=0.6), autopct='%1.1f%%')
ax.axis("equal")
plt.title('Распределение среднего количества рейсов в день в процентном соотношении по городам ТОП-10', y=1.03)
plt.show()
print('ТОП-10 городов составляет {:.1%} от всего количества рейсов'.format(data_2pie_t.average_flights[1:].sum()/data_2pie_t.average_flights.sum()))
```

### Вывод

   На графике отчётливо видно, что ТОП-10 городов составляет 44.8% от всго количества рейсов в города, и из них практически половина рейсов 23.4% направляются в Москву
